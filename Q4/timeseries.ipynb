```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Watches Bags Accessories.csv', encoding='latin-1')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

### Cleaning

```python
df["Rating in Stars"].unique()
df['Rating in Stars'].replace(['5-Apr', '5-May','5-Mar','5-Jan' ], np.nan, inplace=True)
df['Rating in Stars'] = df['Rating in Stars'].str.replace('/5', '').astype(float)
```

```python
df['Sold Count'] = df['Sold Count'].str.replace('Sold', '')
df['Sold Count'] = df['Sold Count'].str.replace('K', '000').str.strip().astype(float)
```

```python
df["Current Price"] = df["Current Price"].str.replace(',', '').astype(float)
```

```python
df["Original Price"] = df["Original Price"].str.replace('Rs.', '')
df["Original Price"] = df["Original Price"].str.replace(',', '').str.strip().astype(float)
```

```python
df.drop(columns=["Currency", "Link"], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

### Impute

```python
group_means = df.groupby('Category')['Rating in Stars'].transform('mean')

df['Rating in Stars'].fillna(group_means, inplace=True)
```

```python
group_means = df.groupby('Category')['Sold Count'].transform('mean')

df['Sold Count'].fillna(group_means, inplace=True)
```

```python
group_means = df.groupby('Category')['Original Price'].transform('mean')

df['Original Price'].fillna(group_means, inplace=True)
```

### Feature Engineering

```python
df["Price Difference"] = df["Original Price"]-df["Current Price"]
df["Total Earning"] = df["Sold Count"]*df["Current Price"]
```
